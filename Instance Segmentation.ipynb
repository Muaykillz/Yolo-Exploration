```python
from ultralytics import YOLO
```

# Define model

```python
model = YOLO("yolov8n-seg.pt")
```

# Train model

```python
results = model.train(data="coco8-seg.yaml", epochs=50, imgsz=640)
```

# Validate model

```python
metrics = model.val()
```

```python
results = model("https://ultralytics.com/images/bus.jpg", save=True)
```

# Export model

```python
model.export(format="onnx")
```
