```python
from ultralytics import YOLO
```

# Define model

```python
model = YOLO("yolov8n-cls.pt")
```

# Train model

```python
results = model.train(data="mnist", epochs=1, imgsz=64)
```

# Validate model

```python
results = model.val()
```

```python
results = model("https://ultralytics.com/images/bus.jpg", save=True)
```

# Export model

```python
model.export(format="onnx")
```
