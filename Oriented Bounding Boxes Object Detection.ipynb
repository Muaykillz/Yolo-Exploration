```python
from ultralytics import YOLO
```

# Define model

```python
model = YOLO("yolov8n-obb.pt")
```

# Train model

```python
results = model.train(data="dota8.yaml", epochs=10, imgsz=640)
```

# Validate model

```python
metrics = model.val(data="dota8.yaml")
```
